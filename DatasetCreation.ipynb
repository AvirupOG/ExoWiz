```python
from typing import Optional, List, Dict, Any, Tuple
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings

import lightkurve as lk

def _safe_get(table, names: List[str], default=None):
    if table is None:
        return default
    colnames = set(getattr(table, "colnames", []))
    lower_map = {c.lower(): c for c in colnames}
    for n in names:
        if n in colnames:
            return np.array(table[n])
        if n.lower() in lower_map:
            return np.array(table[lower_map[n.lower()]])
    return default


def _get_header_vals(hdu, keys: List[str]) -> Dict[str, Any]:
    out = {}
    if hdu is None:
        return {k: None for k in keys}
    hdr = getattr(hdu, "header", None)
    for k in keys:
        out[k] = hdr.get(k) if hdr is not None else None
    return out


def _tolist(x):
    if x is None:
        return None
    if isinstance(x, (np.ndarray, list, tuple)):
        try:
            return x.tolist()
        except Exception:
            return list(x)
    return x  # scalar

def fetch_lightcurves(
    target_id: Any,
    mission: str = "Kepler",
    apply_quality_mask: bool = True,
    use_tpf: bool = False
) -> List[Dict[str, Any]]:
    tgt = str(target_id).strip()
    if mission.lower() == "kepler" and not tgt.upper().startswith(("KIC", "KEPLER")):
        query = f"KIC {tgt}"
    elif mission.lower() == "tess" and not tgt.upper().startswith(("TIC", "TESS")):
        query = f"TIC {tgt}"
    else:
        query = tgt

    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        search = lk.search_lightcurvefile(query, mission=mission)

    if len(search) == 0:
        return []

    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        lcf_collection = search.download_all()

    results: List[Dict[str, Any]] = []

    for lcf in lcf_collection:
        lc_sap_raw = None
        lc_pdc_raw = None
        try:
            lc_sap_raw = lcf.SAP_FLUX
        except Exception:
            pass
        try:
            lc_pdc_raw = lcf.PDCSAP_FLUX
        except Exception:
            pass

        lc_primary_raw = lc_pdc_raw or lc_sap_raw
        if lc_primary_raw is None:
            continue
        if apply_quality_mask and hasattr(lc_primary_raw, "quality"):
            good = (lc_primary_raw.quality == 0)
        else:
            good = slice(None)
        lc_primary = lc_primary_raw[good]
        def _maybe_mask(lc_raw):
            if lc_raw is None:
                return None
            try:
                if isinstance(good, slice) or len(lc_raw) == len(lc_primary_raw):
                    return lc_raw[good]
                return lc_raw
            except Exception:
                return lc_raw

        lc_sap = _maybe_mask(lc_sap_raw)
        lc_pdc = _maybe_mask(lc_pdc_raw)
        try:
            lc_primary = lc_primary.remove_nans()
        except Exception:
            pass
        try:
            if lc_sap is not None:
                lc_sap = lc_sap.remove_nans()
        except Exception:
            pass
        try:
            if lc_pdc is not None:
                lc_pdc = lc_pdc.remove_nans()
        except Exception:
            pass
        if lc_primary is None or len(lc_primary) == 0:
            continue
        tab_primary = lc_primary.to_table()
        tab_sap = lc_sap.to_table() if lc_sap is not None and len(lc_sap) > 0 else None
        tab_pdc = lc_pdc.to_table() if lc_pdc is not None and len(lc_pdc) > 0 else None
        time = np.array(lc_primary.time.value)
        flux_raw = _safe_get(tab_sap or tab_primary, ["SAP_FLUX", "flux"], default=None)
        flux_corr = _safe_get(tab_pdc or tab_primary, ["PDCSAP_FLUX", "flux"], default=None)
        flux_err = _safe_get(tab_primary, ["PDCSAP_FLUX_ERR", "SAP_FLUX_ERR", "flux_err"], default=None)
        quality = _safe_get(tab_primary, ["QUALITY", "quality"], default=None)
        cadenceno = _safe_get(tab_primary, ["CADENCENO", "cadenceno"], default=None)
        sap_bkg = _safe_get(tab_primary, ["SAP_BKG", "sap_bkg"], default=None)
        mom_c1 = _safe_get(tab_primary, ["MOM_CENTR1", "mom_centr1"])
        mom_c2 = _safe_get(tab_primary, ["MOM_CENTR2", "mom_centr2"])
        psf_c1 = _safe_get(tab_primary, ["PSF_CENTR1", "psf_centr1"])
        psf_c2 = _safe_get(tab_primary, ["PSF_CENTR2", "psf_centr2"])
        pos_c1 = _safe_get(tab_primary, ["POS_CORR1", "pos_corr1"])
        pos_c2 = _safe_get(tab_primary, ["POS_CORR2", "pos_corr2"])
        try:
            hdu = lcf.hdu[1]
        except Exception:
            hdu = None

        meta_hdr = _get_header_vals(hdu, [
            "CROWDSAP", "FLFRCSAP", "MISSION", "OBJECT", "RA_OBJ", "DEC_OBJ",
            "TIMEUNIT", "BJDREFI", "BJDREFF", "BJDREF", "TELAPSE"
        ])
        sector = getattr(lcf, "sector", None)
        quarter = getattr(lcf, "quarter", None)
        campaign = getattr(lcf, "campaign", None)
        cadence_min = None
        try:
            tt = lc_primary.time.value
            if len(tt) > 1:
                cadence_min = float(np.nanmedian(np.diff(tt)) * 24.0 * 60.0)
        except Exception:
            pass
        cx_tpf = cy_tpf = None
        if use_tpf:
            try:
                with warnings.catch_warnings():
                    warnings.simplefilter("ignore")
                    tpf_search = lk.search_targetpixelfile(
                        query, mission=mission, sector=sector, quarter=quarter, campaign=campaign
                    )
                if len(tpf_search) > 0:
                    with warnings.catch_warnings():
                        warnings.simplefilter("ignore")
                        tpf = tpf_search.download()
                    flux_cube = tpf.flux.value
                    if flux_cube is not None and flux_cube.ndim == 3 and flux_cube.size > 0:
                        rows = np.arange(flux_cube.shape[1])[:, None]
                        cols = np.arange(flux_cube.shape[2])[None, :]
                        num_y = (flux_cube * rows).sum(axis=(1, 2))
                        num_x = (flux_cube * cols).sum(axis=(1, 2))
                        den = flux_cube.sum(axis=(1, 2)) + 1e-12
                        cy_tpf = num_y / den
                        cx_tpf = num_x / den
            except Exception:
                pass

        results.append({
            "mission": mission,
            "target_query": query,
            "sector": sector,
            "quarter": quarter,
            "campaign": campaign,
            "cadence_min": cadence_min,
            "time": time,
            "cadenceno": cadenceno,
            "flux_raw": flux_raw,
            "flux_corr": flux_corr,
            "flux_err": flux_err,
            "quality": quality,
            "sap_bkg": sap_bkg,
            "mom_centr1": mom_c1, "mom_centr2": mom_c2,
            "psf_centr1": psf_c1, "psf_centr2": psf_c2,
            "pos_corr1": pos_c1, "pos_corr2": pos_c2,
            "cx_tpf": cx_tpf, "cy_tpf": cy_tpf,
            **meta_hdr,
        })

    return results


def fetch_lightcurve(
    target_id: Any,
    mission: str = "Kepler",
    prefer_pdcsap: bool = True
) -> Tuple[Optional[np.ndarray], Optional[np.ndarray]]:
    res = fetch_lightcurves(target_id, mission=mission, apply_quality_mask=True, use_tpf=False)
    if not res:
        return None, None
    first = res[0]
    time = first.get("time")
    flux = first.get("flux_corr") if prefer_pdcsap else first.get("flux_raw")
    return time, flux

def build_vetting_dataset(
    df: pd.DataFrame,
    id_col: str = "kepid",            # your ID column in df
    label_col: Optional[str] = "label",
    mission: str = "Kepler",          # "Kepler" or "TESS"
    apply_quality_mask: bool = True,
    use_tpf: bool = False,
    take_pdcsap: bool = True,
    save_parquet_path: Optional[str] = None,
    save_csv_path: Optional[str] = None,
) -> pd.DataFrame:
    if id_col not in df.columns:
        raise ValueError(f"df must contain '{id_col}'")

    has_label = label_col in df.columns if label_col else False
    iterable = df[[id_col, label_col]].itertuples(index=False, name=None) if has_label \
               else df[[id_col]].itertuples(index=False, name=None)

    records: List[Dict[str, Any]] = []

    for item in tqdm(iterable, total=len(df), desc="Building vetting dataset"):
        if has_label:
            target_id, label = item
        else:
            target_id = item[0]
            label = None

        try:
            results = fetch_lightcurves(
                target_id=target_id,
                mission=mission,
                apply_quality_mask=apply_quality_mask,
                use_tpf=use_tpf
            )
            if not results:
                print(f"[INFO] no LCFs for {target_id}")
        except Exception as e:
            print(f"[WARN] fetch failed for {target_id}: {repr(e)}")
            results = []

        for r in (results or []):
            # choose primary flux
            flux_primary = r.get("flux_corr") if take_pdcsap else r.get("flux_raw")
            if flux_primary is None:
                flux_primary = r.get("flux_corr") or r.get("flux_raw")

            rec = {
                # ids/labels
                "target_id": int(target_id) if str(target_id).isdigit() else str(target_id),
                "label": int(label) if label is not None else None,
                # observing block identity
                "mission": r.get("mission"),
                "sector": r.get("sector"),
                "quarter": r.get("quarter"),
                "campaign": r.get("campaign"),
                "cadence_min": r.get("cadence_min"),
                # series
                "time": _tolist(r.get("time")),
                "flux": _tolist(flux_primary),
                "flux_raw": _tolist(r.get("flux_raw")),
                "flux_corr": _tolist(r.get("flux_corr")),
                "flux_err": _tolist(r.get("flux_err")),
                "quality": _tolist(r.get("quality")),
                "cadenceno": _tolist(r.get("cadenceno")),
                # background / crowding
                "sap_bkg": _tolist(r.get("sap_bkg")),
                "CROWDSAP": r.get("CROWDSAP"),
                "FLFRCSAP": r.get("FLFRCSAP"),
                # centroid / motion (LCF-level)
                "mom_centr1": _tolist(r.get("mom_centr1")),
                "mom_centr2": _tolist(r.get("mom_centr2")),
                "psf_centr1": _tolist(r.get("psf_centr1")),
                "psf_centr2": _tolist(r.get("psf_centr2")),
                "pos_corr1": _tolist(r.get("pos_corr1")),
                "pos_corr2": _tolist(r.get("pos_corr2")),
                # optional TPF simple centroids
                "cx_tpf": _tolist(r.get("cx_tpf")),
                "cy_tpf": _tolist(r.get("cy_tpf")),
                # lineage
                "RA_OBJ": r.get("RA_OBJ"),
                "DEC_OBJ": r.get("DEC_OBJ"),
                "BJDREFI": r.get("BJDREFI"),
                "BJDREFF": r.get("BJDREFF"),
                "BJDREF": r.get("BJDREF"),
                "TELAPSE": r.get("TELAPSE"),
                "target_query": r.get("target_query"),
            }
            if rec["time"] is None or rec["flux"] is None:
                continue

            records.append(rec)

    out = pd.DataFrame.from_records(records)
    if not out.empty:
        out["obs_block_id"] = (
            out["mission"].astype(str) + "-" +
            out["target_id"].astype(str) + "-S" +
            out["sector"].astype(str) + "-Q" +
            out["quarter"].astype(str) + "-C" +
            out["campaign"].astype(str)
        )

    if save_parquet_path:
        out.to_parquet(save_parquet_path, index=False)
    if save_csv_path:
        out.to_csv(save_csv_path, index=False)

    return out

if __name__ == "__main__":
    vet_df = build_vetting_dataset(
        df=df.head(20),
        id_col="kepid",
        label_col="label",
        mission="Kepler",
        apply_quality_mask=True,
        use_tpf=False,
        take_pdcsap=True,
        save_parquet_path="vetting_kepler.parquet",
        save_csv_path=None
    )

    print("Columns:", vet_df.columns.tolist())
    print(vet_df.head(20))
    print(f"Rows: {len(vet_df)}  (targets × observed blocks)")
```

```python
def fetch_lightcurve(kepid):
    try:
        search = lk.search_lightcurvefile(f"KIC {int(kepid)}", mission="Kepler")
        if len(search) == 0:
            return None, None
        lc = search.download().PDCSAP_FLUX.remove_nans()
        return lc.time.value, lc.flux.value
    except Exception as e:
        print(f"Failed for {kepid}: {e}")
        return None, None
```

```python
records = []

for _, row in tqdm(df.iterrows(), total=len(df)):
    time, flux = fetch_lightcurve(row["kepid"])
    if time is not None:
        records.append({
            "kepid": int(row["kepid"]),
            "label": int(row["label"]),
            "time": time,
            "flux": flux
        })

df_raw = pd.DataFrame(records)
print(df_raw.head())
```
